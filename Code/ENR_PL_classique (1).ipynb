```python
import numpy as np
import pandas as pd
import pulp
import holoviews as hv
import panel as pn
hv.extension('bokeh')
```

```python
# Chargement des données
df_solar = pd.read_csv('solar.csv')
df_wind = pd.read_csv('wind_onshore.csv')
df_demand = pd.read_csv('demand2050_ADEME.csv', header=None)
df_demand.columns = ["heures", "demande"]

# Calcul consommation annuelle
annual_demand = df_demand['demande'].sum()

# Calcul production potentielle pour 1GW installé
annual_solar_per_gw = df_solar['facteur_charge'].sum()
annual_wind_per_gw = df_wind['facteur_charge'].sum()

# Pour avoir 100% de surproduction et un mix 50-50
target_production = annual_demand * 3.5
wind_capacity = target_production/(2 * annual_wind_per_gw)
solar_capacity = target_production/(2 * annual_solar_per_gw)

print(f"Capacités requises: éolien = {wind_capacity:.1f}GW, solaire = {solar_capacity:.1f}GW")
```

```python
def optimize_storage(solar_profile, wind_profile, demand, wind_cap, solar_cap,
                    phs_capacity=180, phs_power=9.3, phs_efficiency=0.75,
                    ptg_capacity=125000, ptg_power_in=7.66, ptg_power_out=32.93, 
                    ptg_efficiency=0.4):
    """
    Optimise la stratégie de stockage sur l'année avec contrainte cyclique
    
    Parameters:
    -----------
    solar_profile, wind_profile: séries temporelles des facteurs de charge
    demand: série temporelle de la demande
    wind_cap, solar_cap: capacités installées en GW
    phs_capacity: capacité de stockage PHS en GWh
    phs_power: puissance max PHS en GW
    phs_efficiency: rendement PHS (aller-retour)
    ptg_capacity: capacité de stockage P2G en GWh
    ptg_power_in: puissance max électrolyse en GW  
    ptg_power_out: puissance max reconversion en GW
    ptg_efficiency: rendement P2G (aller-retour)
    """
    
    N = len(demand)  # nombre d'heures
    
    # Création du problème
    prob = pulp.LpProblem("Storage_Strategy", pulp.LpMinimize)
    
    # Variables
    phs_in = pulp.LpVariable.dicts("phs_in", range(N), 0)  # Charge PHS
    phs_out = pulp.LpVariable.dicts("phs_out", range(N), 0)  # Décharge PHS
    phs_level = pulp.LpVariable.dicts("phs_level", range(N), 0)  # Niveau PHS
    
    ptg_in = pulp.LpVariable.dicts("ptg_in", range(N), 0)  # Charge P2G
    ptg_out = pulp.LpVariable.dicts("ptg_out", range(N), 0)  # Décharge P2G
    ptg_level = pulp.LpVariable.dicts("ptg_level", range(N), 0)  # Niveau P2G
    
    deficit = pulp.LpVariable.dicts("deficit", range(N), 0)  # Déficit
    
    # Production ENR horaire
    prod = {t: wind_cap * wind_profile[t] + solar_cap * solar_profile[t] 
            for t in range(N)}
    
    # Fonction objectif : minimiser déficits
    prob += -ptg_level[N-1] + 1e5 * pulp.lpSum(deficit[t] for t in range(N))
    
    # Contraintes
    
    # Niveaux initiaux
    prob += phs_level[0] == phs_capacity/2  # PHS à 50%
    prob += ptg_level[0] == ptg_capacity    # P2G plein
    
    # Contraintes cycliques : les niveaux finaux doivent être égaux aux niveaux initiaux
    #prob += phs_level[N-1] == phs_level[0]
    #prob += ptg_level[N-1] == ptg_level[0]
    
    for t in range(N):
        # Équilibre offre-demande
        prob += prod[t] + phs_out[t]*phs_efficiency + ptg_out[t]*ptg_efficiency - \
                phs_in[t] - ptg_in[t] + deficit[t] >= demand[t]
        
        # Limites PHS
        prob += phs_in[t] <= phs_power
        prob += phs_out[t] <= phs_power
        prob += phs_level[t] <= phs_capacity
        
        # Limites P2G
        prob += ptg_in[t] <= ptg_power_in
        prob += ptg_out[t] <= ptg_power_out
        prob += ptg_level[t] <= ptg_capacity
        
        # Évolution des stocks
        if t > 0:
            prob += phs_level[t] == phs_level[t-1] + phs_in[t]*phs_efficiency - phs_out[t]
            prob += ptg_level[t] == ptg_level[t-1] + ptg_in[t]*ptg_efficiency - ptg_out[t]
       
        
    # Résolution
    prob.solve()
    
    return {
        'status': pulp.LpStatus[prob.status],
        'phs_in': [phs_in[t].value() for t in range(N)],
        'phs_out': [phs_out[t].value() for t in range(N)],
        'phs_level': [phs_level[t].value() for t in range(N)],
        'ptg_in': [ptg_in[t].value() for t in range(N)],
        'ptg_out': [ptg_out[t].value() for t in range(N)],
        'ptg_level': [ptg_level[t].value() for t in range(N)],
        'deficit': [deficit[t].value() for t in range(N)]
    }
```

```python
# Lancer l'optimisation
result = optimize_storage(
    df_solar['facteur_charge'].values,
    df_wind['facteur_charge'].values,
    df_demand['demande'].values,
    wind_cap=wind_capacity,
    solar_cap=solar_capacity
)

# Analyse des résultats
print(f"Status: {result['status']}")
print(f"Déficit total: {sum(result['deficit']):.1f} GWh")
print(f"PHS final: {result['phs_level'][-1]:.1f} GWh")
print(f"P2G final: {result['ptg_level'][-1]:.1f} GWh")

```
